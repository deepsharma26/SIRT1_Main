```python
from padelpy import padeldescriptor

fingerprint = 'MACCS'

fingerprint_output_file = ''.join([fingerprint,'.csv'])
fingerprint_descriptortypes = fp[fingerprint]

padeldescriptor(mol_dir='molecule.smi',
                d_file=fingerprint_output_file,

                descriptortypes= fingerprint_descriptortypes,
                detectaromaticity=True,
                standardizenitro=True,
                standardizetautomers=True,
                threads=1,
                removesalt=True,
                log=True,
                fingerprints=True)
```

```python
descriptors = pd.read_csv(fingerprint_output_file)
descriptors
```

```python

original_df = pd.read_csv('/content/npass_data_Batch1.csv')


descriptors_df = pd.read_csv(fingerprint_output_file)


final_df = pd.concat([original_df[['SMILES']], descriptors_df], axis=1)

#
final_output_file = f"{fingerprint}_with_SMILES_Batch1.csv"
final_df.to_csv(final_output_file, index=False)

print(f"Final file saved as {final_output_file}")
```
