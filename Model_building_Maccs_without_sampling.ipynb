#Buiding LazyClassfier model without any sampling strategies.
We have selected the fatures in three different categories: Top10, Top20, and Top30. Furthermore, with the imbalanaced dataset, we have developed the best model to confirm the best algroithm and best feature category.

```python
!pip install lazypredict
import pandas as pd
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier
import warnings
warnings.filterwarnings("ignore")

# --- Dictionary to store results ---
model_results = {}

# --- List of files and labels ---
files = [
    ('/content/MACCS2_Trainset_top10.csv', 'Top 10 Features'),
    ('/content/MACCS2_Trainset_top20.csv', 'Top 20 Features'),
    ('/content/MACCS2_Trainset_top30.csv', 'Top 30 Features')
]

all_results = []  # to store combined results

# --- Loop through each dataset ---
for filepath, label in files:
    print(f"\n🔍 Running LazyClassifier on: {label} (With imbalanced dataset)")

    # Load dataset
    df = pd.read_csv(filepath)
    df = df.drop('Unnamed: 0', axis=1, errors='ignore')

    X = df.drop("BioactivityClass", axis=1)
    Y = df["BioactivityClass"]

    # Train-test split (no SMOTE)
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.3, stratify=Y, random_state=100
    )

    # LazyClassifier on original (imbalanced) data
    clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
    models, predictions = clf.fit(X_train, X_test, Y_train, Y_test)

    # Save and print results
    model_results[label] = models
    print(models)

    # Save individual results to CSV
    models.to_csv(f"LazyClassifier_results_{label.replace(' ', '_')}.csv")

    # Append dataset label for merged summary
    models_reset = models.reset_index().rename(columns={"index": "Model"})
    models_reset.insert(0, "Feature_Set", label)
    all_results.append(models_reset)

# --- Save combined results into one CSV ---
final_df = pd.concat(all_results, ignore_index=True)
final_df.to_csv("LazyClassifier_all_results.csv", index=False)

print("\n✅ All results saved:")
print("   - Individual CSVs per dataset")
print("   - Merged file: LazyClassifier_all_results.csv")
```
